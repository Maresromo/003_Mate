```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

## y' = y*sin(2t), y(0) = 1, 0<=t<=2pi

```python
def f(t,y0):
    return y0*(math.sin(2*t))     #math.sin() trabaja en radianes
```

## y'' = y'sen(2t) + 2y*cos(2t)

```python
def fprima(t,y0):
    return f(t,y0)*(math.sin(2*t)) + 2*y0*math.cos(2*t)
```

```python
def y_analitica(t,y0,t0):
    c = y0*math.exp(1/2*math.cos(2*t0))
    y = c/(math.exp(1/2*math.cos(2*t)))
    return y
```

```python
#Aca se definen las funciones de k1 hasta k4 necesarias para el met de Runge Kutta
# como k1 = f 

def k1(t,y0):
    return f(t,y0)
    
def k2(t,y0,h):
    a = t + 1/2*h
    b = y0 + 1/2*k1(t,y0)*h
    return f(a,b)                        #k2 = f(a,b)  lol hice por partes para que sea mas facil de visualizar

def k3(t,y0,h):
    a = t + 1/2*h
    b = y0 + 1/2*k2(t,y0,h)*h
    return f(a,b)                        #k3 = f(a,b)  lol hice por partes para que sea mas facil de visualizar

def k4(t,y0,h):
    a = t + h
    b = y0 + k3(t,y0,h)*h
    return f(a,b)                        #k4 = f(a,b)  lol hice por partes para que sea mas facil de visualizar

def Runge_Kutta(t,y0,h):
    return y0 + 1/6*h*(k1(t,y0) + 2*k2(t,y0,h) + 2*k3(t,y0,h) + k4(t,y0,h))
```

```python
#En este caso no se uso la solucion analitica, solo metodos numericos

t0 = 0.
h = 0.001        #Delta de tiempo
tmax = 2*math.pi

y0 = y1 = y2 = y3 = y4 = 1

t_values = np.arange(t0, tmax+h, h)  #Esto es la parte del tiempo, nos permite tener mas precision que si lo incluyeramos como una suma en el for

yE = []                           #Lista vacia donde se guardaran los datos del metodo de Euler
yT = []                           #Lista vacia donde se guardaran los datos del metodo de Taylor de segundo orden
yRK = []                          #Lista vacia donde se guardaran los datos del metodo de Runge Kutta de cuarto orden
yA = []

for t in t_values:
    
    yE.append(y1)              #Metodo de Euler, lo uinico que se tiene que cambiar es la funcion f
    y1 = y1 + f(t,y1)*h         # la cual es la derivada con la que estamos trabajando

    yT.append(y2)                            #En el met de Taylor de segundo orden, necesitamos la primera y segunda derivada
    y2 = y2 + h*f(t,y2) + h*h/2*fprima(t,y2)

    yRK.append(y3)                            #En el met de Runge Kutta de cuart orden utilizamos la funcion definida arriba
    y3 = Runge_Kutta(t,y3,h)

    yA.append(y4)
    y4 = y_analitica(t,y0,t0)

p4 = pd.DataFrame({            #Aca se integran los valores de cada metodo en un DF
    'tk':t_values,
    'yE':yE,
    'yT':yT,
    'yRK':yRK,
    'yA':yA
})

p4
```

```python
desde = t0
hasta = tmax
p4_filtered = p4[(p4['tk'] >= desde) & (p4['tk'] <= hasta)]

plt.plot(p4_filtered['tk'], p4_filtered['yA'], label='Analitico', linestyle='-', color='yellow', marker='x', markersize=4, linewidth=5, alpha=0.8)
plt.plot(p4_filtered['tk'], p4_filtered['yT'], label='Taylor', linestyle='--', color='green', marker='s', markersize=4, linewidth=5, alpha=0.8)
plt.plot(p4_filtered['tk'], p4_filtered['yRK'], label='Runge-Kutta', linestyle='-.', color='red', marker='x', markersize=4, linewidth=5, alpha=0.8)
plt.plot(p4_filtered['tk'], p4_filtered['yE'], label='Euler', linestyle='-', color='blue', marker='o', markersize=4, linewidth=5)

plt.xlabel('t')
plt.ylabel('y')
plt.title('''005a Comparacin de Mtodos Numricos 
f(y) = y*sin(2t), y(0) = 1, 0<=t<=2pi''')
plt.legend()
plt.grid(True)
plt.savefig('005a_Grafico_comparacion.jpg', format='jpg', dpi=150)  # dpi ajusta la calidad
plt.show()
```
