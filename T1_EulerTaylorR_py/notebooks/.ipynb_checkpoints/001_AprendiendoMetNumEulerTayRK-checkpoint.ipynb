```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def y(t):
    return 0.5 * (math.exp(2 * t) + 1)
```

```python
tmax = 1
y0 = 1.
delta = 0.001
t = 0.
t_values = np.arange(t, tmax+delta, delta)

tk = []
yk = []
ya = []

for t in t_values:
    tk.append(t)
    yk.append(y0)
    ya.append(y(t))
    y1 = y0 + (2*y0 -1)*delta
    y0 = y1

Euler1 = pd.DataFrame({
    'tk':tk,
    'yk':yk,
    'ya':ya
})

Euler1
```

## 2. y' = k*y*(1-y)

```python
def f(k,y0):
    return k*y0*(1-y0)
```

## y'' = k(1-2y)y'

```python
def fprima(k,y0):
    return k*(1-2*y0)*f(k,y0)
```

```python
#En este caso no se uso la solucion analitica, solo metodos numericos

t = 0.
h = 0.01         #Delta de tiempo
tmax = 6

k = 1.
y1 = y2 = 0.5

t_values = np.arange(t, tmax+h, h)  #Esto es la parte del tiempo, nos permite tener mas precision que si lo incluyeramos como una suma en el for

yE = []                           #Lista vacia donde se guardaran los datos del metodo de Euler de segundo orden
yT = []                           #Lista vacia donde se guardaran los datos del metodo de Taylor de segundo orden

for t in t_values:
    
    yE.append(y1)              #Metodo de Euler, lo uinico que se tiene que cambiar es la funcion f
    y1 = y1 + f(k,y1)*h         # la cual es la derivada con la que estamos trabajando

    yT.append(y2)                            #En el met de Taylor de segundo orden, necesitamos la primera y segunda derivada
    y2 = y2 + h*f(k,y2) + h*h/2*fprima(k,y2)

p2 = pd.DataFrame({            #Aca se integran los valores de cada metodo en un DF
    'tk':t_values,
    'yE':yE,
    'yT':yT,
})

p2
```
